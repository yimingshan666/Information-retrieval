### Task 2
For this task, download the modified version of the *Yale Face Database B* provided on Moodle (`task2_data.zip`). The Yale Face Database B consists of single light source images of $10$ subjects, each seen in different poses and illumination conditions. In the provided form the database is divided into $5$ subsets. In subset $0$ the subject is illuminated by an almost frontal light source, while for subsets $1$-$4$ the light source is gradually moved along the horizon. Subset $0$ will serve as the training set, while subsets $1$-$4$ are used for testing.
#### *Part A*
Write a function that takes as an input matrix $\mathbf{T}$ of vectorized images from subset $0$. The output of this function are the $20$ first singular vectors $\mathbf{U}[:, 1]\,,\;...\,,\;\mathbf{U}[:, 20]$. Display the first $3$ vectors as images, i.e., reshape them to size $50 \times 50$ and display them.

```python
import imageio
import numpy as np
from os import listdir
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
def read_data(path, array_of_img, array_of_label, img_h, img_w):
    
    for filename in listdir(r"./" + path):
        img = np.reshape(np.array(imageio.imread(path + "/" + filename)), (img_h * img_w)) / 255
        array_of_img.append(img)
        array_of_label.append(int(filename[6: 8]))
    # centralize data (X)    
    X = np.array(array_of_img).T  # shape=(p, N)
    array_of_img = X - np.mean(X, axis = 1).reshape(X.shape[0], 1)  # shape=(p, N)

    return array_of_img, np.array(array_of_label)
```

```python
# Block for Data Loading
subset_id = [0, 1, 2, 3, 4]
img_h = 50
img_w = 50
data_dir = 'yaleBfaces'
subsets = {}
labels = {}

for idx in subset_id:
    subset_name = 'subset' + str(idx)

    subsets[idx] = []  # X
    labels[idx] = []  # y

    subsets[idx], labels[idx] = read_data(data_dir + '/' + subset_name, subsets[idx], labels[idx], img_h, img_w)
```

```python
def principle_comp_getter(T):
    '''
    T: Training set
    k: num of principle components
    
    return first k sigular vectors of SVD(T_centered)
    '''

    U, Sigma, VT = np.linalg.svd(T, full_matrices = False)

    return U[:, 0: 20]
```

```python
T = subsets[0]

U = principle_comp_getter(T)
```

```python
plt.figure(figsize=(20, 20))
plt.subplot(1,3,1)
plt.imshow(U[:,0].reshape(50, 50), cmap='gray')
plt.title('Singular vector (0)')
plt.ylabel('y')
plt.xlabel('x')

plt.subplot(1,3,2)
plt.imshow(U[:,1].reshape(50, 50), cmap='gray')
plt.title('Singular vector (1)')
plt.ylabel('y')
plt.xlabel('x')

plt.subplot(1,3,3)
plt.imshow(U[:,2].reshape(50, 50), cmap='gray')
plt.title('Singular vector (2)')
plt.ylabel('y')
plt.xlabel('x')


plt.show()
```

#### *Part B*
Write a function that takes as an input the training set $\mathbf{T}$ (a matrix composed of vectorized pictures from subset $0$), a vector containing the labels of the training set (i.e., if the $i$-th sample belongs to class $j$, the $i$-th entry of the labels vector is $j$. In this exercise $j$ is an integer between $1$ and $10$), the test samples $S$ (a matrix composed of vectorized pictures from subsets $1$-$4$) and the corresponding labels (in a separate vector), the $20$ singular vectors from the first step, and the parameter $k$ that denotes how many of the PCs are used. Use the Euclidean distance to classify each sample image based on its three nearest neighbors (this is done by comparing the test samples with the training samples in the reduced space.). As an output give the fraction of images from $S$ that were misclassified, i.e., the error rate. Repeat this for subsets $1$ through $4$ and for $k=1\,,\;...\,,\;20$. Plot the error rate for each subset.

```python
# define function for error calculation
def err_knn_pca(T, T_labs, S, S_labs, U, k, test_flag=False):
    '''
    T: training set / S:testing set （T/S is p x N matrix）
    T_labs/S_labs is the labels of T and S（T_labs / S_labs is size N vector）
    U is the eigenvector after svd（U is p x k matrix）
    k is the number of igenvectors we want（k is scalar value）
    test_flag for special test, used in part C
    
    return error rate
    '''
    model_knn = KNN(n_neighbors = 3)
    U_k = U[:, 0: k]
    # Apply PC transformation for dimensionality reduction
    U_k_T = np.transpose(U_k)
    T_new = np.dot(U_k_T, T)

    # Train KNN on T_new
    model_knn.fit(np.transpose(T_new), T_labs)
    
    S_new = np.dot(U_k_T, S)
    # Test KNN on S_new
    if test_flag and (k == 3 or k == 4):
        print("Prediction of image person_02_16 in subset 1: ", model_knn.predict(np.transpose(S_new))[20])
        print("Groundtruth of image person_02_16 in subset 1: ", S_labs[20])
    acc = model_knn.score(np.transpose(S_new), S_labs)

    return 1 - acc
```

```python
# calculate errors for each subset and k 
k_num = range(1, 21)
err = {}
subset_num = [0, 1, 2, 3, 4]

T = subsets[0]
T_labs = labels[0]

# the loop for all subsets
for i in subset_num[1:]:

    err[i] = np.zeros(len(k_num))
    
    # calculate error as k increases

    for k in k_num:
        S = subsets[i]
        S_labs = labels[i]
        err[i][k-1] = err_knn_pca(T, T_labs, S, S_labs, U, k)
```

```python
for j in subset_num[1:]:
    plt.plot(k_num, err[j], label='subset ' + str(j))

plt.title('Error rate for subsets 1-4 and k from 1-20')
plt.xlabel('k')
plt.ylabel('Error rate')
plt.grid()
plt.legend()

plt.show()
```

#### *Part C*
Repeat the above experiment without using the first three singular vectors, i.e., use $k=1\,,\;...\,,\;17$ singular vectors starting from the $4$-th. Plot the error rate as before. How do you explain the difference in recognition rate?

```python
# modify U to remove first three columns of k
U_modified = U[:, 3:]

# calculate errors for each subset and k 

k_num_modified = range(1, 18)
err_modified = {}

T = subsets[0]
T_labs = labels[0]

# iteration among all test subsets
for i in subset_num[1:]:

    err_modified[i] = np.zeros(len(k_num_modified))
    # calculate error while k increases
    for k in k_num_modified:
        S = subsets[i]
        S_labs = labels[i]
        err_modified[i][k-1] = err_knn_pca(T, T_labs, S, S_labs, U_modified, k)
```

```python
f, ax = plt.subplots(len(subset_num[1:]), sharex=True, figsize=(10,20))


for i in subset_num[1:]:
    ax[i-1].plot(k_num, err[i], color='blue', label='original $U_k$')
    ax[i-1].plot(k_num_modified, err_modified[i], color='red', label='First three cols of $U_k$ removed')
    ax[i-1].set_title('Error rate for subset ' + str(i) + ' as number of used componets varies')
    ax[i-1].set_xlabel('k')
    ax[i-1].set_ylabel('Error rate')
    ax[i-1].grid()
    ax[i-1].legend()


    
plt.show()
```

In our case, $\mathbf{X  =USV}^\top$, and Score Matrix $\mathbf{S  =U^\top * USV}^\top$, where $\mathbf{U}$ indicates the principal directions and Score Matrix $\mathbf{S}$ represents the principal components.


**For Subset 1 and Subset 2**, it can be observed that the model with original $\mathbf{U}$ is doing better when the number of used components is small. As the number k increases, both perform almost at the same level. 

But, when k = 3 or 4 for subset 1, the model with original $\mathbf{U}$ performs worse. We print some information (slot below) based on the image person_02_16 which has **much noise**. The model with original $\mathbf{U}$ misclassified the image all the time and on the contrary, the model with first 3 cols removed $\mathbf{U}$ always predict correctly. Our guess is, noised data's most important principal components **might** also be biased due to the noise and then it leads to misclassification.


**For subsets 3**, removing the first three cols of $\mathbf{U}$ results in better performance. The reason might be: After projecting the principal components to the subspace of $\mathbf{U}$, the most important 3 principal components capture more variance, and thus data with different labels have many overlaps, and this hurts the classification model's performance.



**For the last subset 4**, error rates of both are too high and after having a look at the data, many images can't be classified by human eyes, so we think it cannot explain anything.
